```python
DF1E = pd.DataFrame({'NOMBRE':['Jose','Max'],
                    'CARRERA':['Economia','Arquitectura'],
                    'EDAD':[23,26]}).set_index('NOMBRE')
```

```python
DF2E = pd.DataFrame({'NOMBRE':['Lupe','Maxi'],
                    'CARRERA':['Medicina','Informatica'],
                    'EDAD':[40,42]}).set_index('NOMBRE')
```

```python
#Concatenar cuando se tienen las mismas columnas y solo más registros
#Concatenacion por filas
DFCOM = pd.concat([DF1E,DF2E])
```

```python
DFCOM
```

```python
DF3E = pd.DataFrame({'NOMBRE':['Jose','Max'],
                    'CARRERA':['Economia','Arquitectura']}).set_index('NOMBRE')
```

```python
DF4E = pd.DataFrame({'NOMBRE':['Lupe','Maxi'],
                    'TURNO':['Matutino','Vespertino']}).set_index('NOMBRE')
```

```python
DFCOM2 = pd.concat([DF3E,DF4E], axis=1)
#Concatenacion por columnas
```

```python
DFCOM2
```

```python
#Mezclar dataframes
```

```python
DF5E = pd.DataFrame({'NOMBRE':['Jose','Max'],
                    'CARRERA':['Economia','Arquitectura']})
```

```python
DF6E = pd.DataFrame({'NOMBRE':['Carmen','Max'],
                    'TURNO':['Matutino','Vespertino']})
```

```python
DFCOM3 = pd.merge(DF5E,DF6E, on='NOMBRE', how='outer')
#outer combinación a través de sus filas y inner por defecto
DFCOM3
```
